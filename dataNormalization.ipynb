## Why we normalize databases

The reason why we normalize databases, is to reduce data redundancy and improve data integrity. Imagine a large production database that, but we are migrating our datastore and we notice some corrupted data and need to update some values on the database. Instead of running a script through every row of the database to detect whether or not it is the right condition to apply a change to the data, we can update the data in one place due to database normalization.

### 1NF - First Normal Form

In order to satisfy 1NF, values in each row must be atomic and should not contain composite or multi-valued attributes i.e. no nested tables.

```python
accountMapping = pd.read_csv(os.path.join(path,csvList[8]))
accountMapping.head()
```

```python
accounts = pd.read_csv(os.path.join(path, csvList[1]))
accounts.head()
```

```python
transactions = pd.read_csv(os.path.join(path, csvList[-1]))
transactions.head()
```

### 2NF - Second Normal Form

In order to satisfy 2NF, any relation must be in 1NF, and there should not be any partial dependencies, i.e. in any given table, no columns should depend on another column that is not the prime attribute. A table in 2NF is said to be fully functionally dependent

```python
accounttype = accounts[[accounts.columns[i] for i in [2,3,4]]].drop_duplicates()
accounttype.to_csv("./6NF/accountType.csv", index=False)
accounttype.head()
```

```python
allAccounts = accounts[[accounts.columns[i] for i in [11,6,0,5,7,8,9,10]]]
allAccounts.to_csv("allAccounts.csv", index=False)
assert(allAccounts.shape == allAccounts.drop_duplicates().shape)
allAccounts.head()
```

```python
kycinfo = accounts[list(accounts.columns[0:2]) + list(accounts.columns[-12:])]
kycinfo.to_csv("kycinfo.csv", index=False)
assert(kycinfo.shape == kycinfo.drop_duplicates().shape)
kycinfo.head()
```

### 3NF - Third Normal Form

In order to satisfy 3NF, we ensure that there are no transitive dependencies i.e. if X->Y and Y->Z, then X->Z; and we must ensure Z is not in the same table

### 4NF Fourth Normal Form

Table should not have any multi-valued dependency in 4NF, i.e. each key should only determine one row.

### 5NF Fifth Normal Form

A table is said to be in the 5NF if and only if every non-trivial join dependency in that table is implied by the candidate keys.

### 6NF Sixth Normal Form

A table is in 6NF if there is only one other attribute other than the key.

```python
priorSar = allAccounts[[allAccounts.columns[i]
                        for i in [0,1,2,3]]].drop_duplicates()
priorSar.to_csv("./6NF/priorSar.csv", index=False)
priorSar.head()
```
